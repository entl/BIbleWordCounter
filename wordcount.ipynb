```python
import matplotlib.pylab as plt
from prettytable import PrettyTable
```

```python
text = open("bible.txt", "r").read()
text = replacement(text).lower().split(" ")
words = {word: text.count(word) for word in set(text)}
words = dict(sorted(words.items(),key=lambda item:item[1], reverse=True))
# print(words)
plot(words)
print(stat_output(words))
```

```python
def stat_output(words):
    t = PrettyTable(['â„–','Word', 'Count'])
    # print(words.items())
    i = 1
    for key, value in words.items():
        t.add_row([i,key,value])
        i+=1
    return(t)
     
```

```python
def plot(words):
    x, y = zip(*words.items())
    plt.xticks(rotation=60, fontsize=11)
    plt.bar(x[:30],y[:30])
    plt.show()
```

```python
def replacement(string):
    chars = "0123456789:;,.()\t\n"
    for char in chars:
        string = string.replace(char, "")
    return string
```
